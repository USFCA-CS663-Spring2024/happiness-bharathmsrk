*  Are there any interesting distributions in the (numeric) features?

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_columns = ['year', 'Life Ladder', 'Log GDP per capita', 'Social support',
                   'Healthy life expectancy at birth', 'Freedom to make life choices',
                   'Generosity', 'Perceptions of corruption', 'Positive affect',
                   'Negative affect']
```

```python
plt.figure(figsize=(15, 15))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(4, 3, i)
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

```python
df["Life Ladder"].skew()
```

```python
df["Life Ladder"].kurt()
```

```python
df["Perceptions of corruption"].skew()
```

```python
df["Perceptions of corruption"].kurt()
```

```python
df["Generosity"].skew()
```

```python
df["Generosity"].kurt()
```

* From the above graphs, we can infer that all the graphs are normal distribution with some of them being left or right skewed. For example, Life Ladder looks the most normal amongst all. While the negative skewness (-1.4904) and positive kurtosis (1.8142) values suggest that the distribution of the "Perceptions of corruption" column is left-skewed and leptokurtic. And the positive skewness (0.7771) and positive kurtosis (0.8316) values suggest that the distribution of the "Generosity" column is right-skewed and leptokurtic.

```python
sns.pairplot(df)
```

```python
# Example of scatter plot along with a regression line to indicate the linear relationship

life_ladder_data = df['Life Ladder'].dropna()
log_gdp_data = df['Log GDP per capita'].dropna()

life_ladder_data = life_ladder_data[log_gdp_data.index]
log_gdp_data = log_gdp_data[life_ladder_data.index]

plt.figure(figsize=(10, 6))
sns.scatterplot(x=log_gdp_data, y=life_ladder_data, color="skyblue", alpha=0.7, label='Data Points')

sns.regplot(x=log_gdp_data, y=life_ladder_data, scatter=False, color="black", line_kws={'label':'Linear Regression'})

plt.title('Scatter Plot: Life Ladder vs. Log GDP per capita')
plt.xlabel('Log GDP per capita')
plt.ylabel('Life Ladder')
plt.legend()
plt.show()
```

There are several pairplots that show somewhat linear relationships.

# Answer to Question 2

```python
# What is the happiest country in 2022?

happiest_country_22 = df[df['year'] == 2022].loc[df[df['year'] == 2022]['Life Ladder'].idxmax(), 'Country name']
print(f"The happiest country in 2022 is: {happiest_country_22}")
```

```python
# What is the happiest country in 2013?

happiest_country_13 = df[df['year'] == 2013].loc[df[df['year'] == 2013]['Life Ladder'].idxmax(), 'Country name']
print(f"The happiest country in 2013 is: {happiest_country_13}")
```

```python
# What is the happiest country of all time?

happiest_country = df.loc[df['Life Ladder'].idxmax(), 'Country name']
print(f"The happiest country of all time is: {happiest_country}")
```

```python
# What is the least happy country in 2022?

least_happy_country_22 = df[df['year'] == 2022].loc[df[df['year'] == 2022]['Life Ladder'].idxmin(), 'Country name']
print(f"The least happy country in 2022 is: {least_happy_country_22}")
```

```python
# What is the least happy country in 2013?

least_happy_country_13 = df[df['year'] == 2013].loc[df[df['year'] == 2013]['Life Ladder'].idxmin(), 'Country name']
print(f"The least happy country in 2013 is: {least_happy_country_13}")
```

```python
# What is the least happy country of all time?

least_happy_country = df.loc[df['Life Ladder'].idxmin(), 'Country name']
print(f"The least happy country of all time is: {least_happy_country}")
```

# Part 3 - 
# Answer to Question 3

```python
# Country with the largest increase in happiness

df = df.sort_values(by=["Country name", "year"])

first_year_happiness = df.groupby("Country name")["Life Ladder"].first()
last_year_happiness = df.groupby("Country name")["Life Ladder"].last()

percentage_change = ((last_year_happiness - first_year_happiness) / first_year_happiness) * 100

max_change_country = percentage_change.idxmax()
max_change_value = percentage_change.max()

print(f"Country with the largest percentage increase in happiness is {max_change_country} with an increase of {max_change_value:.2f}%.")
```

```python
# Country with the largest decrease in happiness

max_decrease_country = percentage_change.idxmin()
max_decrease_value = percentage_change.min()

print(f"The country with the largest decrease in happiness is {max_decrease_country} with a decrease of {max_decrease_value:.2f}.")
```
